```python
import os
import Klubplywacki

swim_files = os.listdir(Klubplywacki.FOLDER)
swim_files.remove(".DS_Store")
swimmers = []
for file in swim_files:
    swimmers.append((Klubplywacki.read_swim_data(file)[0], file))

print(swimmers[:6])  # Print first 6 swimmers for verification

names = []
for row in swimmers:
    if row[0] not in names:
        names.append(row[0])
print(sorted(names))
```
